# Playing with Meta-SNE and Common CNN Models

First we need to import some libraries.

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.autograd import Variable
from torchvision import models
from torchvision import transforms

import PIL
from PIL import Image

from os import listdir

from sklearn.manifold import TSNE

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
```

We also need to load the image data.

```python
imsize = 224
root = 'ilsvrc_parse/sel_images/'
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
loader = transforms.Compose([transforms.Scale((imsize, imsize)),
                             transforms.ToTensor(),
                             normalize])
image_names = listdir(root)

def image_loader(image_name):
    image = Image.open(image_name)
    image = Variable(loader(image))
    return image



images = [image_loader(root + img_name).data for img_name in image_names]

images = np.array([img for img in images 
                   if img.numpy().shape == (3,224,224)])

np.random.seed(2)
indices = np.random.choice(len(images), 100)
images = images[indices]

# for img in images:
#    print(img.numpy().shape)

images = Variable(torch.stack((images)))
```

Sanity check.

```python
def imshow_noax(img, normalize=True):
    """ Tiny helper to show images as uint8 and remove axis labels """
    if normalize:
        img_max, img_min = np.max(img), np.min(img)
        img = 255.0 * (img - img_min) / (img_max - img_min)
    plt.imshow(img.astype('uint8'))
    plt.gca().axis('off')

img = images[0].data.numpy()
img = np.moveaxis(img, 0, -1)
print(img.shape)
imshow_noax(img)
plt.show()
```

Next, we load some pretrained ImageNet models. For now we're going to see if we can just get the visualization for AlexNet working.

```python
alexnet = models.alexnet(pretrained=True)
densenet = models.densenet161(pretrained=True)
resnet = models.resnet18(pretrained=True)
squeezenet = models.squeezenet1_0(pretrained=True)
vgg = models.vgg16(pretrained=True)

nets = [alexnet]
```

Now we need to run the 100 images through the network and store the outputs of the layers we want.

```python
modlist = torch.nn.ModuleList()

def param_count(module):
    total = 0
    for p in module.parameters():
        total += p.data.numpy().size
    return total

for net in nets:
    for m in list(net.modules()):
#        if param_count(m) < 1e4:
        if True:
            modlist.append(m)

# 2 5 8 10 12 17 20 22
#layer_indices = {'Conv2d':[2,5,8,10,12],
#                 'ReLU':[3,6,9,11,13,18,21],
#                 'MaxPool2d':[4,7,14],
#                 'Linear':[17,20,22]}
# all_indices = np.sort(np.concatenate(list(layer_indices.values())))
all_indices = np.arange(len(modlist))
print(all_indices)
layer_outputs = []

def c(self, input, output):
    layer_outputs.append(output)
    
for i in all_indices:
    modlist[i].register_forward_hook(c)
```

```python
len(images)
```

We can now run the forward pass through all of the images and all of the networks

```python
chunk_size = 10
image_chunks = int(len(images) / chunk_size)
for net in nets:
    for i in range(image_chunks):
        print('Chunk ' + str(i))
        net.forward(images[chunk_size*i:chunk_size*(i+1)])
```

```python
print(len(layer_outputs))
```

```python
layer_chunks = int(len(layer_outputs) / chunk_size)
shortened_layer_outputs = []
for i in range(layer_chunks):
    shortened_layer_outputs.append(torch.cat(layer_outputs[i::layer_chunks]))
layer_outputs = shortened_layer_outputs
```

Next we'll define the function used to construct the distance matrices.

```python
def distance_matrix(outputs):
    '''
    outputs -- ordered list of representations for all images
    '''
    distances = []
    for i in range(100):
        for j in range(i+1, 100):
            dist = torch.dist(outputs[i], outputs[j]).data.numpy()
            distances.append(dist)
    return np.concatenate(distances)
```

Sanity checking the representations.

```python
representations = np.stack([distance_matrix(output) for output in layer_outputs])
for r in representations:
    print(len(r))
    print(r)
```

Now we run t-SNE on the representations.

```python
embedded = TSNE(n_components=2, verbose=1, perplexity=30, learning_rate=500, n_iter=5000, method='exact', init='pca').fit_transform(representations)
X, Y = embedded[:,0], embedded[:,1]
print(embedded)

indices = {'Conv2d':[0,3,6,8,10],
           'ReLU':[1,4,7,9,11,14,16],
           'MaxPool2d':[2,5,12],
           'Linear':[13,15,17]}

colors = ['black', 'black', 'blue', 'pink', 'red', 'blue', 'pink', 'red', 'blue', 'pink', 'blue', 'pink', 
          'blue', 'pink', 'red', 'black', 'yellow','green','pink','yellow','green','pink','green']

plt.scatter(X[indices['Conv2d']],Y[indices['Conv2d']], c='red')
plt.scatter(X[indices['ReLU']],Y[indices['ReLU']], c='blue')
plt.scatter(X[indices['MaxPool2d']],Y[indices['MaxPool2d']], c='orange')
plt.scatter(X[indices['Linear']],Y[indices['Linear']], c='purple')
#plt.scatter(X,Y)
plt.show()
```
